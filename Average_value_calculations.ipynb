```python
import os
import pandas as pd
import numpy as np

def parse_xvg(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            if line.startswith(('@', '#')):
                continue
            parts = line.strip().split()
            if len(parts) >= 2:
                try:
                    time = float(parts[0])
                    value = float(parts[1])
                    data.append((time, value))
                except ValueError:
                    continue
    return pd.DataFrame(data, columns=['Time', 'Value'])

def summarize_xvg(directory='.', pattern='rmsd', output_csv="rmsd_summary.csv"):
    summary = []
    for filename in os.listdir(directory):
        if filename.endswith(".xvg") and pattern in filename.lower():
            file_path = os.path.join(directory, filename)
            df = parse_xvg(file_path)
            if not df.empty:
                avg = df['Value'].mean()
                std = df['Value'].std()
                summary.append({
                    'Filename': filename,
                    'Average': avg,
                    'Standard Deviation': std
                })

    summary_df = pd.DataFrame(summary)
    summary_df.to_csv(output_csv, index=False)
    print(f"Summary saved to: {output_csv}")
    return summary_df

# === Run summary function for all RMSD files ===
results = summarize_xvg()
results
```

```python
import os
import pandas as pd

# Function to read RMSF .xvg file
def parse_rmsf_xvg(file_path):
    data = []
    with open(file_path, 'r') as file:
        for line in file:
            if line.startswith(('@', '#')):
                continue
            parts = line.strip().split()
            if len(parts) >= 2:
                try:
                    time = float(parts[0])
                    value = float(parts[1])
                    data.append((time, value))
                except ValueError:
                    continue
    return pd.DataFrame(data, columns=['Time', 'Value'])

# Summary function for RMSF
def summarize_rmsf():
    summary = []
    for file in os.listdir():
        if file.endswith('.xvg') and 'rmsf' in file.lower():
            df = parse_rmsf_xvg(file)
            if not df.empty:
                avg_val = df['Value'].mean()
                std_val = df['Value'].std()
                summary.append({
                    'Filename': file,
                    'Average RMSF': avg_val,
                    'Standard Deviation': std_val
                })
    result_df = pd.DataFrame(summary)
    result_df.to_csv("rmsf_summary.csv", index=False)
    print(" RMSF summary saved to 'rmsf_summary.csv'")
    return result_df

# Run the RMSF summary function
rmsf_summary_df = summarize_rmsf()
rmsf_summary_df
```

```python
import os
import numpy as np
import pandas as pd

# Path to your uploaded files
path = '/content'
# Adjust this filter if needed (e.g., 'gyrate' or 'rg' in name)
rg_files = [f for f in os.listdir(path) if f.endswith('.xvg') and 'gyrate' in f.lower()]

results = []

for file in rg_files:
    rg_total_values = []

    with open(os.path.join(path, file), 'r') as f:
        for line in f:
            if line.startswith(('@', '#')):
                continue  # Skip header/comments
            parts = line.strip().split()
            if len(parts) >= 2:
                try:
                    rg_total = float(parts[1])  # s0: total Rg
                    rg_total_values.append(rg_total)
                except ValueError:
                    continue

    if rg_total_values:  # Ensure we got valid numbers
        avg_rg = np.mean(rg_total_values)
        std_rg = np.std(rg_total_values)

        results.append({
            'Filename': file,
            'Average_Rg (nm)': round(avg_rg, 4),
            'SD_Rg (nm)': round(std_rg, 4)
        })

# Create DataFrame
df_rg = pd.DataFrame(results)
print(df_rg)

# Save summary
df_rg.to_csv('Rg_summary.csv', index=False)
print(" Rg summary saved to 'Rg_summary.csv'")
```
